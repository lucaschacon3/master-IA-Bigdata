# Tipos de IA: Débil General (AGI) y Fuerte

## IA Débil (Narrow AI)
Es la única forma de inteligencia artificial existente hoy. Está diseñada para realizar **una tarea específica** o un conjunto muy limitado de funciones. No posee comprensión general ni conciencia.

- **Características**:
  - Especializada en un dominio concreto.
  - Supera al humano en su tarea, pero no puede aplicarse a otros campos.
  - Aprende a partir de datos (machine learning, deep learning).

- **Ejemplos prácticos**:
  - **Asistentes virtuales**: Siri, Alexa (reconocen voz y ejecutan comandos, pero no “entienden” el mundo).
  - **Sistemas de recomendación**: Netflix, YouTube, Spotify (personalizan contenido, pero no hacen otra cosa).
  - **Detección de fraude**: modelos bancarios que analizan transacciones en milisegundos.
  - **Vehículos autónomos**: Tesla Autopilot (conducen, pero no pueden mantener una conversación).
  - **AlphaGo**: juega Go a nivel experto, pero no puede jugar al ajedrez sin reprogramarse.
  - **LLMs actuales** (GPT-4, etc.): aunque parecen generalistas, siguen siendo IA débil porque están limitadas a procesar lenguaje; no poseen comprensión real ni autonomía fuera de ese ámbito.

- **Casos de uso**:
  - **Medicina**: diagnóstico de imágenes (radiología, patología) con precisión similar a especialistas.
  - **Retail**: control de stock con sensores IoT que predicen demanda y reordenan productos.
  - **Telecomunicaciones**: predicción de **churn** (clientes que se van a la competencia) para ofrecer retenciones personalizadas.
  - **Industria**: robots que realizan soldadura, ensamblaje o inspección con visión por computadora.

---

## IA General (AGI – Artificial General Intelligence)
Es un concepto aún hipotético. Se define como una IA con **capacidad cognitiva equivalente o superior a la humana** en un amplio abanico de dominios. Podría aprender, razonar, planificar, entender contexto y adaptarse a problemas nunca antes vistos sin necesidad de reentrenamiento.

- **Características**:
  - Generaliza conocimiento entre dominios.
  - Posee sentido común, comprensión causal, y capacidad de transferir habilidades.
  - No existe actualmente (aunque algunos modelos avanzados muestran destellos).

- **Estado actual**:
  - Los **LLM** actuales (GPT, Gemini) son a menudo confundidos con AGI porque manejan lenguaje general, pero siguen siendo IA débil: no razonan de manera consistente, alucinan hechos y no tienen comprensión real del mundo.
  - Muchos investigadores consideran que estamos a décadas de lograr una AGI verdadera.

- **Ejemplos hipotéticos**:
  - Un agente capaz de cocinar, planificar un viaje, escribir código, reparar un coche y conversar sobre filosofía, todo con la misma “mente”.
  - Sistemas que puedan adaptarse a cualquier empleo humano sin reprogramación específica.

- **Casos de uso potenciales**:
  - Automatización completa de cualquier trabajo cognitivo.
  - Investigación científica autónoma (descubrimiento de teorías, diseño de experimentos).
  - Gestión de crisis complejas (cambios climáticos, pandemias) integrando múltiples disciplinas.

---

## IA Fuerte (Strong AI)
Es un concepto filosófico: la IA fuerte no solo actuaría como humana, sino que **poseería conciencia, sentimientos y subjetividad**. Es el concepto de una “mente artificial” con experiencias propias.

- **Características**:
  - **Consciencia** y **capacidad de sentir**.
  - Se debate si es posible o si la consciencia es un fenómeno biológico irreproducible.

- **Estado actual**:
  - Totalmente especulativo. No existe evidencia de que las máquinas actuales tengan experiencia subjetiva.

---

## Resumen práctico

| Tipo | Estado | Capacidad | Ejemplo real | Caso de uso actual |
|------|--------|-----------|--------------|-------------------|
| **IA Débil** | Existente | Específica, no generaliza | Siri, AlphaGo, GPT-4, sistema de fraude | Asistentes, recomendación, diagnóstico, control de stock, chatbots |
| **AGI** | Hipotética (no lograda) | Generaliza como humano | Ninguno aún | Automatización de cualquier trabajo cognitivo |
| **IA Fuerte** | Filosófico (no lograda) | Conciencia + inteligencia general | Ninguno | Máquinas con derechos, compañía consciente |
