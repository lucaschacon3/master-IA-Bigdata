
## Ética, Legalidad y Gobernanza en IA

### 1. Ética y Legalidad
La **ética** en IA busca que los sistemas sean explicables, justos (sin sesgos) y robustos. La **legalidad** transforma estos principios en normas vinculantes. Un sistema puede ser legal pero no ético, o viceversa; el objetivo del cumplimiento (compliance) es alinear ambos.

### 2. Privacy by Design
Es un enfoque preventivo donde la protección de datos se integra en el **diseño del sistema** desde su concepción, no como un añadido posterior.
* **Minimización:** Solo recolectar los datos estrictamente necesarios.
* **Anonimización/Pseudonimización:** Técnicas para que los datos no identifiquen a individuos.
* **Seguridad Extremo a Extremo:** Protección del dato durante todo su ciclo de vida.

### 3. Peligros y Casos Reales
El despliegue de IA sin control genera riesgos críticos:
* **Sesgo Algorítmico:** El caso de *Amazon* y su herramienta de contratación que discriminaba a mujeres por datos históricos sesgados.
* **Alucinaciones y Desinformación:** Modelos que generan hechos falsos con apariencia de verdad (Deepfakes).
* **Violación de Copyright:** Uso de obras protegidas para entrenar modelos sin consentimiento (casos actuales de artistas contra Midjourney/OpenAI).

### 4. EU AI Act y AESIA
* **EU AI Act:** Es la primera ley integral de IA en el mundo (Unión Europea). Clasifica los sistemas por **niveles de riesgo**:
    * *Riesgo Inaceptable:* (Ej. puntuación social) → Prohibidos.
    * *Alto Riesgo:* (Ej. salud, justicia, infraestructuras) → Requisitos estrictos de transparencia y supervisión.
    * *Riesgo Limitado:* (Ej. chatbots) → Obligación de informar que se interactúa con una IA.
    * *Riesgo Mínimo:* (Ej. detector de Spam) → No obligación de informar que se interactúa con una IA.

* **AESIA:** Agencia Española de Supervisión de la Inteligencia Artificial. Es el organismo encargado en España de velar por el cumplimiento del EU AI Act, siendo pionera en Europa.

### 5. Shadow AI
Se refiere al uso de herramientas de IA dentro de una organización **sin la aprobación o el conocimiento** del departamento de IT o Seguridad.
* **Peligro:** Empleados subiendo datos confidenciales de la empresa a ChatGPT u otros modelos públicos, provocando fugas de información masivas.

### 6. Gobernanza y Compliance
* **Gobernanza:** El conjunto de políticas, roles y procesos que definen cómo se crea y usa la IA en una entidad. Establece quién es responsable de cada decisión del modelo.
* **Compliance (Cumplimiento):** El proceso técnico y legal para asegurar que el sistema cumple con las leyes (GDPR, AI Act) y los estándares internos.
