# Subcampos de la Inteligencia Artificial

La IA abarca múltiples disciplinas, cada una con enfoques y aplicaciones específicas. A continuación se describen los más relevantes en la actualidad.

---

## 1. IA (Inteligencia Artificial) – Término general
Es el campo amplio que busca crear sistemas capaces de realizar tareas que normalmente requieren inteligencia humana. Incluye todos los demás subcampos.

**Ejemplos**: desde un chatbot hasta un coche autónomo.  
**Casos de uso**: automatización de procesos, análisis de datos, toma de decisiones.

---

## 2. ML (Machine Learning – Aprendizaje Automático)
Subcampo donde las máquinas aprenden patrones a partir de datos sin ser programadas explícitamente. Se divide en supervisado, no supervisado y por refuerzo.

**Ejemplos**:  
- Modelos que predicen el precio de una vivienda (supervisado).  
- Segmentación de clientes (no supervisado).  
- AlphaGo (refuerzo).

**Casos de uso**:  
- Detección de fraude en transacciones bancarias.  
- Predicción de **churn** en telecomunicaciones.  
- Sistemas de recomendación (Netflix, YouTube).

---

## 3. DL (Deep Learning – Aprendizaje Profundo) y Redes Neuronales
Subcampo del ML basado en redes neuronales de múltiples capas (profundas). Permite procesar datos no estructurados como imágenes, audio y texto con un rendimiento excepcional.

**Ejemplos**:  
- Redes neuronales convolucionales (CNN) para reconocimiento de imágenes.  
- Redes recurrentes (RNN) y transformers para secuencias de texto.

**Casos de uso**:  
- Diagnóstico médico por imágenes (radiología, patología).  
- Conducción autónoma (percepción de objetos).  
- Asistentes de voz (Siri, Alexa).

---

## 4. GenAI (IA Generativa)
Subcampo centrado en crear contenido nuevo (texto, imágenes, audio, vídeo, código) a partir de patrones aprendidos. Utiliza principalmente modelos como LLM (grandes modelos de lenguaje) y modelos de difusión.

**Ejemplos**:  
- ChatGPT (texto).  
- Midjourney, DALL·E (imágenes).  
- Suno (música).  
- GitHub Copilot (código).

**Casos de uso**:  
- Generación de contenidos de marketing.  
- Prototipado rápido de software.  
- Creación de asistentes conversacionales avanzados.  
- Síntesis de datos para entrenar otros modelos.

---

## 5. IA Agéntica (Agentic AI)
Enfoque donde los sistemas actúan como **agentes autónomos** que persiguen objetivos, toman decisiones y ejecutan acciones en entornos dinámicos, a menudo con poca o ninguna supervisión humana. Combina LLM, planificación, memoria y herramientas externas.

**Ejemplos**:  
- Agentes que reservan citas por teléfono.  
- Agentes de código que desarrollan software completo.  
- Robots autónomos que navegan en almacenes.

**Casos de uso**:  
- Automatización de flujos de trabajo complejos.  
- Atención al cliente con agentes que resuelven incidencias sin intervención humana.  
- Asistentes personales que coordinan agenda, compras y viajes.

---

## 6. NLP (Natural Language Processing – Procesamiento de Lenguaje Natural)
Subcampo dedicado a que las máquinas comprendan, interpreten y generen lenguaje humano. Es la base de asistentes, chatbots y análisis de texto.

**Ejemplos**:  
- Análisis de sentimiento en redes sociales.  
- Traducción automática (Google Translate).  
- Extracción de información de documentos legales.

**Casos de uso**:  
- Chatbots de soporte.  
- Clasificación de tickets de atención al cliente.  
- Resumen automático de documentos.  
- Detección de **síntomas médicos** en conversaciones con pacientes.

---

## 7. CV (Computer Vision – Visión por Computadora)
Subcampo que permite a las máquinas “ver” e interpretar el mundo visual: imágenes y vídeos.

**Ejemplos**:  
- Reconocimiento facial.  
- Conteo de personas con drones o cámaras fijas.  
- Inspección de calidad en fábricas.

**Casos de uso**:  
- Drones para vigilancia y conteo de personas.  
- Cajas autónomas (Amazon Go).  
- Diagnóstico de imágenes médicas (detección de tumores).  
- Vehículos autónomos (detección de peatones, señales).

---

## 8. Sistemas Expertos
Uno de los primeros enfoques de IA. Utilizan reglas lógicas y bases de conocimiento codificadas por expertos humanos para resolver problemas en dominios específicos. A diferencia del ML, no aprenden de datos, sino que aplican reglas predefinidas.

**Ejemplos**:  
- MYCIN (diagnóstico de enfermedades infecciosas, años 70).  
- Sistemas de configuración de equipos complejos.

**Casos de uso actuales**:  
- Sistemas de soporte a decisiones en áreas muy reguladas (cumplimiento normativo).  
- Motores de reglas en finanzas para autorización de operaciones.  
- Chatbots con flujos estructurados (cuando no hay datos suficientes para ML).

---

## 9. Robótica e IoT (Internet of Things)
La **robótica** combina IA con hardware para que los robots interactúen físicamente con el mundo. El **IoT** conecta sensores y dispositivos, y junto con la IA permite la automatización y el análisis en tiempo real.

**Ejemplos**:  
- Robots de limpieza (Roomba).  
- Drones de inspección.  
- Sensores de temperatura en almacenes conectados a IA que predicen averías.

**Casos de uso**:  
- **Robots en hogar**: aspiradoras, cortacéspedes autónomos.  
- **Robótica industrial**: brazos robotizados con visión por computadora.  
- **IoT + IA**: agricultura de precisión (sensores que optimizan riego), ciudades inteligentes (tráfico, iluminación), wearables médicos que alertan ante anomalías.

---

## Relación entre subcampos (práctica)

| Subcampo | Se apoya en | Aplicación típica |
|----------|-------------|------------------|
| NLP | ML, DL, GenAI | Chatbots, asistentes |
| CV | ML, DL, GenAI | Drones, diagnóstico médico |
| GenAI | DL (transformers, difusión) | Creación de contenido |
| IA Agéntica | NLP, CV, GenAI, robótica | Agentes autónomos |
| Robótica + IoT | CV, ML, sistemas expertos | Automatización industrial y doméstica |

---

**Nota práctica**:  
En la industria actual, la mayoría de las soluciones combinan varios subcampos. Por ejemplo, un asistente virtual usa **NLP** (entender la pregunta), **GenAI** (generar respuesta), **ML** (personalizar), y puede actuar como **agente** si ejecuta acciones como encender luces (integrado con IoT).
