# Potencial y desafíos de la Inteligencia Artificial

## Potencial: lo que la IA puede hacer por nosotros

- **Automatización de tareas repetitivas**: en retail, sistemas de control de stock que predicen la demanda y reordenan productos automáticamente.

- **Democratización del conocimiento**: un pequeño negocio puede usar asistentes virtuales para atención al cliente sin necesidad de contratar un equipo.

- **Mejora en la toma de decisiones**: predicción de **churn** (clientes que se van a la competencia) para ofrecer retenciones personalizadas; **detección de fraude** con modelos que identifican transacciones anómalas en milisegundos.

- **Avances científicos y médicos**: en **medicina**, análisis de imágenes médicas (radiología, patología) con precisión comparable a la de especialistas; en **descubrimiento de fármacos**, la IA reduce tiempos de investigación; **sensores y wearables** (IoT) monitorizan pacientes crónicos y alertan ante anomalías.

- **Personalización masiva**: **sistemas de recomendación** como Netflix, YouTube o Spotify adaptan contenido a gustos individuales; en educación, rutas de aprendizaje adaptativas según el ritmo del estudiante.

- **Nuevos modelos de negocio y servicios**: **agentes autónomos** como asistentes (Siri, Alexa), chatbots y drones de reparto; **IAs generativas** que crean contenido (texto, imagen, vídeo) bajo demanda.


## Desafíos: lo que debemos gestionar con cuidado

### 1. Sesgos y discriminación
- Los modelos aprenden de datos históricos que pueden contener sesgos humanos.
- **Ejemplo práctico**: sistemas de selección de personal que discriminaban por género o raza si no se auditan.
- **Mitigación**: auditorías de equidad, datos más diversos, transparencia algorítmica.

### 2. Falta de transparencia y explicabilidad
- Los **LLM** y redes profundas actúan como “cajas negras”: no siempre sabemos por qué dan una respuesta.
- **Ejemplo práctico**: un modelo rechaza un crédito, pero no puede explicar los motivos al usuario.
- **Necesidad**: desarrollar **IA explicable (XAI)** para cumplir normativas (como el GDPR).

### 3. Privacidad y seguridad
- La IA requiere grandes cantidades de datos, a menudo personales.
- **Riesgos**: filtraciones, reidentificación de datos anonimizados, vigilancia masiva.
- **Ejemplo práctico**: uso de cámaras con reconocimiento facial sin consentimiento.

### 4. Impacto laboral y desigualdad
- Algunos empleos se automatizan; se requiere recualificación.
- **No competimos con la IA, sino con personas que usan IA**.
- **Ejemplo práctico**: diseñadores gráficos que usan herramientas generativas aumentan su productividad; quienes no, pueden quedar rezagados.

### 5. Regulación y marcos éticos
- La tecnología avanza más rápido que las leyes.
- **Desafío**: crear normativas que fomenten la innovación sin comprometer derechos.
- **Ejemplo práctico**: el **AI Act** europeo clasifica los sistemas por riesgo y prohíbe ciertos usos (como puntuación social) y la AESIA (Agencia Española de Supervisión de la Inteligencia Artificial).

### 6. Sostenibilidad y coste energético
- Entrenar grandes modelos consume electricidad equivalente a cientos de hogares.
- **Ejemplo práctico**: se investigan modelos más eficientes y uso de energías renovables en centros de datos.

### 7. Coste computacional
- Para entrenar grandes modelos se necesitan datacenter con una gran potencia computacional de GPU, Ra,.
- **Ejemplo práctico**: Empresas como OpenAI, Anthropic, Google, Mistral, etc dan acceso a sus modelos vía API, muchos de estos modelos están en servidores hiperescaleres (AWS, Azure, GCP).

### 8. Dependencia tecnológica y concentración de poder
- Pocas empresas controlan la infraestructura, los datos y los modelos más avanzados.
- **Riesgo**: falta de soberanía tecnológica para países y regiones.

### 9. Desinformación y contenido sintético
- **IAs generativas** permiten crear deepfakes, noticias falsas, reseñas fraudulentas.
- **Ejemplo práctico**: vídeos manipulados de políticos o celebridades que parecen reales.
