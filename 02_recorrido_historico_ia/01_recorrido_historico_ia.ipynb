# Recorrido histórico de la IA

## Las grandes revoluciones
La humanidad ha vivido tres grandes revoluciones que cambiaron la forma de producir, vivir y pensar:
- **Revolución agrícola** (neolítico): domesticación de plantas y animales → primeras civilizaciones.
- **Revolución industrial** (s. XVIII-XIX): máquina de vapor, producción mecanizada → fábricas, urbanización.
- **Revolución electrónica y digital** (s. XX-XXI): computadoras, internet, digitalización → transformación de la información.

## La tercera revolución digital y la IA
La revolución digital tiene una tercera ola: **la inteligencia artificial**.  
Sus pilares:
- **Democratiza el conocimiento**: hoy cualquiera puede acceder a información y herramientas avanzadas (ej. ChatGPT, asistentes en el móvil).
- **Aumenta la eficiencia y productividad**: automatiza tareas repetitivas, analiza datos a gran escala, permite tomar decisiones más rápidas.
- La IA puede **reemplazar tareas**, pero **no competimos con ella**.
- El enfoque práctico es usarla como **herramienta de aumento**: el profesional + IA rinde más que cada uno por separado.

## Pioneros del pensamiento computacional
Antes de la IA moderna, hubo mentes que sentaron las bases:
- **Ramon Llull** (s. XIII): intentó crear un sistema mecánico para generar conocimiento (primer "motor de razonamiento").
- **René Descartes** (s. XVII): visión del cuerpo como máquina, racionalismo y método cartesiano.
- **George Boole** (s. XIX): álgebra booleana → base lógica de los circuitos digitales. True (1) or False(0).
- **Alan Turing** (s. XX): teoría de la computación, prueba del concepto de "máquina universal", y el **test de Turing** para evaluar inteligencia.

## Hitos clave en la historia de la IA

| Hito | Año (aprox.) | Descripción práctica |
|------|--------------|----------------------|
| **Concepto de robot** | 1920 | Obra *R.U.R.* de Čapek: acuña el término "robot". |
| **Algoritmo y computadora** | 1930-1940 | Turing, von Neumann: la computadora programable hace posible la IA. |
| **Asimov y las 3 leyes** | 1942 | Isaac Asimov propone leyes éticas para robots, influyendo en la robótica real. |
| **Conferencia de Dartmouth** | 1956 | Nace oficialmente el campo de la "inteligencia artificial". |
| **Primera red neuronal** | 1958 | Perceptrón de Rosenblatt: inspira el aprendizaje automático actual. |
| **ELIZA (primer chatbot)** | 1966 | Simula una psicoterapeuta; antepasado de los asistentes modernos. |
| **Asistentes virtuales y smartphone** | 2007-2011 | **iPhone** (2007) + **Siri** (2011): la IA llega al bolsillo. **Alexa** (2014) populariza los agentes de voz en el hogar. |
| **DeepBlue vs. AlphaGo** | 1997 / 2016 | DeepBlue (ajedrez) vence a Kasparov; AlphaGo (Go) vence a Lee Sedol – hito en aprendizaje por refuerzo. |
| **Robots en hogar** | 2000s-2020s | Aspiradoras Roomba, asistentes de voz, pequeños robots sociales. |
| **IAs generativas** | 2017-2023 | **LLM** (grandes modelos de lenguaje) como GPT: generan texto, imágenes, código. Revolución en creatividad y automatización. |

## ¿Qué es realmente la IA?
- **No es inteligente** en el sentido humano.
- **Nos imita** a partir de enormes cantidades de datos generados por nosotros.
- Es un **sistema estadístico** que encuentra patrones y los reproduce con variaciones.

## Campos de aplicación práctica (con ejemplos reales)

| Campo | Aplicación concreta | Conceptos integrados |
|-------|---------------------|----------------------|
| **Asistentes virtuales** | Siri, Alexa, Google Assistant | **Agentes**, **LLM**, **iPhone**, **smartphone** |
| **Generación de presupuestos** | Herramientas que analizan historial de ventas y generan ofertas automáticas | **LLM**, automatización |
| **Sistemas de recomendación** | Netflix, YouTube, Spotify | Algoritmos de filtrado colaborativo – “porque viste…” |
| **Control de stock / retail** | Amazon Go, predicción de demanda | **IoT** (sensores en estanterías), visión por computadora |
| **Sector bancario** | **Detección de fraude** en tarjetas, aprobación de créditos | Modelos de anomalías, aprendizaje supervisado |
| **Telecomunicaciones** | Predicción de **churn** (clientes que se van a la competencia) | Análisis de comportamiento, alertas tempranas |
| **Medicina** | Diagnóstico asistido: análisis de imágenes médicas, **síntomas médicos** con chatbots | **LLM**, redes neuronales, sensores (IoT médico) |
| **Drones y análisis de video** | **Conteo de personas**, vigilancia, inspección de infraestructuras | **IoT**, visión por computadora, sensores embarcados |
