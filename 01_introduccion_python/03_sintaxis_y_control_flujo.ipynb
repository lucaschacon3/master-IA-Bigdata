### Variables: Constantes, Locales y Globales

- **Variables**: Se crean al asignarles un valor. No necesitan declaración.

```python
nombre = "Ana"
nombre: str = "Ana" # Se puede usar type hint

print(nombre)
```

* **Constantes:** Por convención (no hay una regla estricta), se nombran en mayúsculas para indicar que no deberían cambiar.

```python
PI = 3.1416

print(PI)
```

- **Variables Locales**: Se definen dentro de una función y solo existen allí.
- **Variables Globales**: Se definen fuera de cualquier función. Se pueden **leer** desde una función pero para **modificarlas**, se usa la palabra clave `global`.

```python
contador = 10

def incrementar():
    global contador
    contador += 1
    print(f"Acceso desde la función:{contador}")

incrementar()
print(f"Acceso desde el exterior:{contador}")
```

### La función `main`

A diferencia de Java, Python no requiere un método `main` para ejecutarse. El código se ejecuta de arriba a abajo. Sin embargo, es una práctica común definir una función `main` para controlar el flujo del programa:

```python
def main():
    print("¡Hola, mundo!")

if __name__ == "__main__":
    main()
```

`if __name__ == "__main__"` verifica si el script se está ejecutando directamente (no importado como módulo) y, en ese caso, llama a `main()`.

### Declaración y Tipado (Dinámico vs Estático)

Python es de **tipado dinámico**. Aunque desde Python 3.6+ se puede usar **type hints** (anotaciones de tipo), estas son solo informativas y no se imponen en tiempo de ejecución.

```python
# Sin type hint (tradicional)
def saludar(nombre):
    return "Hola " + nombre

# Con type hint (mejor para legibilidad)
def saludar(nombre: str) -> str:
    return "Hola " + nombre
```

### Tipos Básicos y Operadores

- **Tipos**:
    - `int` : Números enteros.
    - `float` : Números decimales, son con “.” no con “,”.
    - `str` : Se usan con `''` o con`""` , son inmutables (Ej. cuando concatenas es un nuevo objeto).
    - `bool`: Los valores pueden ser `True` o `False` .
    - `None` : Valores nulos, para comprobar si una variable es None se usa `is` -> `if x is None:` .
- **Para castear:**
        

```python
variable = 1.0

print("Sin casteo:", type(variable))
print("Con casteo a cadena:", type(str(variable)))
print("Con casteo a entero:", type(int(variable)))
print("Con casteo a flotante:", type(float(variable)))

variable= "123"
print("Sin casteo:", type(variable))
print("Con casteo a entero:", type(int(variable)))

# Castaea cuando tiene sentido, int("Hola") -> Error
variable= "hola"
print("Sin casteo:", type(variable))
print("Con casteo a cadena:", type(int(variable)))
```

- **Operadores**:
    - Aritméticos (`+`, , , `/`, `//`, `%`, )
    - Comparación (`==`, `!=`, `<`, `>`, `<=`, `>=`): Ej. comparar cadenas y ordenar alfabéticamente, no como en Java.
    - Lógicos (`and`, `or`, `not`): No se usan caracteres como `&` o `|`.
    - Asignación (`=`, `+=`, etc.): El concepto `variable++`, en Python no existe.
    - Pertenencia (`in`, `not in`): Ej. Para listas, como el `.contains()` en Java.
    - Identidad (`is`, `is not`): Compara el objeto, no el valor.

### Estructuras de Control

```python
# Condicional if-elif-else
edad = 18
if edad >= 18:
    print("Mayor de edad")
elif edad > 0:
    print("Menor de edad")
else:
    print("Edad no válida")

print("----------")

# Bucle for (itera sobre secuencias)
for i in range(5):  # range(inicio, fin, incremento) -> range(0,5,1)
    print(i)   

print("----------")     

for letra in "Python":
    print(letra)

print("----------")

# Bucle while
contador = 0
while contador < 5:
    print(contador)
    contador += 1
```

### Funciones

Se definen con `def`. Pueden tener argumentos por defecto y retornar múltiples valores.

```python
def suma(a, b=1):  # b es opcional, por defecto 1
    """Esta función suma dos números."""  # comentario Docstring
    return a + b

resultado = suma(5, 3) 
print("Resultado de la suma:", resultado)

suma_con_un_numero = suma(5)  # b toma el valor por defecto 0
print("Resultado de la suma con un número:", suma_con_un_numero)
```

### Excepciones

Son errores que ocurren durante la ejecución del programa y que interrumpen el flujo normal. Python proporciona mecanismos para “capturar” y manejar estos errores de forma controlada.

### **Bloques `try`, `except`, `else`, `finally`**

```python
try:
    # Código que puede lanzar una excepción
    numero = int(input("Introduce un número: "))
    resultado = 10 / numero
    print(f"Resultado:{resultado}")
except ValueError:
    # Se ejecuta si ocurre un error de valor (ej. letra en lugar de número)
    print("Error: Debes introducir un número válido.")
except ZeroDivisionError:
    # Se ejecuta si se intenta dividir por cero
    print("Error: No se puede dividir entre cero.")
except Exception as e:
    # Captura cualquier otra excepción (genérico, no muy recomendado)
    print(f"Error inesperado:{e}")
else:
    # Se ejecuta si NO ocurrió ninguna excepción en el bloque try
    print("Todo funcionó correctamente.")
finally:
    # Se ejecuta SIEMPRE (haya o no excepción)
    print("Bloque finally: limpieza o cierre de recursos.")
```

### **Excepciones comunes**

- `ValueError`: cuando una función recibe un argumento de tipo correcto pero valor inapropiado.
- `TypeError`: operación aplicada a objeto de tipo incorrecto.
- `IndexError`: índice fuera de rango (listas, tuplas).
- `KeyError`: clave no encontrada en diccionario.
- `FileNotFoundError`: archivo no encontrado.
- `ZeroDivisionError`: división por cero.
- `ImportError`: fallo al importar un módulo.

### **Lanzar excepciones con `raise`**

Puedes provocar intencionadamente una excepción usando `raise`.

```python
def dividir(a, b):
    if b == 0:
        raise ValueError("El divisor no puede ser cero")
    return a / b

try:
    resultado = dividir(10, 0)
except ValueError as e:
    print(e)
```

### **Crear excepciones personalizadas**

Heredando de `Exception` (o de cualquier subclase).

```python
class SaldoInsuficienteError(Exception):
    """Excepción lanzada cuando no hay suficiente saldo."""
    pass

def retirar_dinero(saldo, cantidad):
    if cantidad > saldo:
        raise SaldoInsuficienteError(f"Saldo insuficiente:{saldo} <{cantidad}")
    return saldo - cantidad

retirar_dinero(100, 150)
```
