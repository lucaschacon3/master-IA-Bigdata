### Clases y Objetos

### **1. Definición básica de una clase**

Una clase es una plantilla para crear objetos. En Python, se define con la palabra clave `class`.


```python
class Perro:
    # Variable de clase (compartida por todas las instancias)
    especie = "Canis familiaris"

    def __init__(self, nombre, edad): # Método constructor
        # Variables de instancia
        self.nombre = nombre
        self.edad = edad

    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau!")

mi_perro = Perro("Rex", 5)
print(mi_perro.nombre)  # Acceso a variable de instancia
print(mi_perro.especie)  # Acceso a variable de clase
mi_perro.ladrar()  # Llamada a método de instancia
```

### **Características**

- **Variables de clase**: se definen fuera de cualquier método y son compartidas por todas las instancias.
- **`__init__`**: es el método constructor (inicializador). Se llama automáticamente al crear una instancia. El primer parámetro es `self`, que referencia a la instancia actual. No es obligatorio, pero es casi indispensable en el 99% de los casos.
- **Variables de instancia**: se crean con `self.nombre`, son propias de cada objeto.
- **Métodos**: funciones definidas dentro de la clase; el primer parámetro es `self`.

En Python, la gestión de visibilidad y acceso a datos se basa en **convenciones** y **decoradores**, no en restricciones estrictas del compilador como en Java.

#### 1. Públicos (Sin guiones)
Es el estándar. Cualquier atributo definido como `self.nombre` es accesible y modificable desde fuera de la clase.
* **Uso:** Para datos que no requieren validación previa.

#### 2. Protected (`_` un guion bajo)
Se define como `self._nombre`. **No impide el acceso**, pero es un aviso para otros programadores: *"Este atributo es de uso interno o para subclases, no lo toques desde fuera"*.
* **Realidad:** Es puramente decorativo; el intérprete te dejará modificarlo, pero rompes la etiqueta del lenguaje.

#### 3. Private (`__` doble guion bajo)
Se define como `self.__nombre`. Activa el **Name Mangling** (deformación de nombre). Python renombra el atributo internamente a `_Clase__nombre` para que sea difícil acceder a él por accidente.
* **Objetivo:** Evitar colisiones de nombres en la herencia, no seguridad infranqueable.

#### 4. Getters y Setters (`@property`)
En lugar de métodos ruidosos como `get_edad()`, usamos el decorador `@property`. Esto permite que un método se use como si fuera un atributo normal, pero ejecutando lógica por detrás.


```python
class Producto:
    def __init__(self, precio):
        self.__precio = precio  # "Privado"

    @property
    def precio(self):
        """Getter: Se activa al hacer 'print(obj.precio)'"""
        return f"{self.__precio}€"

    @precio.setter
    def precio(self, valor):
        """Setter: Se activa al hacer 'obj.precio = 10'"""
        if valor < 0:
            raise ValueError("Precio no válido")
        self.__precio = valor

producto = Producto(20)
print(producto.precio)  # Llama al getter
producto.precio = 25  # Llama al setter
print(producto.precio)  # Llama al getter nuevamente
```

### **3. El parámetro `self`**

- `self` es una referencia a la instancia actual. No es una palabra reservada (podrías llamarlo `this`), pero **por convención se usa `self`**.
- En Java, `this` es implícito; en Python hay que declararlo explícitamente como primer parámetro.

```python
class Ejemplo:
    def metodo(self, x):
        self.x = x   # Asigna a la instancia
```

### **4. Variables de clase vs de instancia**

```python
class Perro:
    especie = "Canis familiaris"   # Variable de clase

    def __init__(self, nombre):
        self.nombre = nombre        # Variable de instancia

# Acceso
print(Perro.especie)                # Desde la clase
print(mi_perro.especie)             # Desde la instancia (hereda el valor)
mi_perro.especie = "Lobo"           # Crea una variable de instancia que oculta la de clase
print(Perro.especie)                # Sigue siendo "Canis familiaris"
print(mi_perro.especie)             # Ahora es "Lobo" (variable de instancia)
```

### **5. Métodos de instancia, de clase y estáticos**

Python ofrece tres tipos de métodos:

```python
class Ejemplo:
    def metodo_instancia(self):
        return "Instancia"

    @classmethod
    def metodo_clase(cls):
        return "Clase"

    @staticmethod
    def metodo_estatico():
        return "Estático"
    
e = Ejemplo()
print(e.metodo_instancia())   # Instancia
print(Ejemplo.metodo_clase()) # Clase (también desde instancia)
print(Ejemplo.metodo_estatico()) # Estático
```

| **Tipo** | **Decorador** | **Primer parámetro** | **Acceso** |
| --- | --- | --- | --- |
| **Instancia** | ninguno | `self` | Puede acceder a la instancia y a la clase |
| **Clase** | `@classmethod` | `cls` | Puede acceder solo a la clase (no a la instancia) |
| **Estático** | `@staticmethod` | ninguno | No accede ni a instancia ni a clase (función dentro de la clase) |

### **6. Herencia**

```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def hablar(self):
        pass   # Método abstracto (en realidad no, pero se sobreescribe)

class Perro(Animal):
    def hablar(self):
        return f"{self.nombre} dice: Guau"

class Gato(Animal):
    def hablar(self):
        return f"{self.nombre} dice: Miau"
```

```python
class Perro(Animal):
    def __init__(self, nombre, raza):
        super().__init__(nombre)   # Llama al constructor de Animal
        self.raza = raza
```

### **`super()`**

- `super()` devuelve una referencia a la clase padre.
- Útil para llamar a métodos de la superclase (constructor, métodos sobrescritos, etc.).

### **Herencia múltiple**

Python permite herencia múltiple:

```python
class A:
    def metodo(self):
        print("A")

class B:
    def metodo(self):
        print("B")

class C(A, B):
    pass

c = C()
c.metodo()   # ¿Qué imprime? Depende del MRO (Method Resolution Order)
print(C.__mro__)  # Muestra el orden de resolución
```

### **7. Métodos especiales (mágicos)**

Los métodos que empiezan y terminan con doble guion bajo (`__` ) son métodos especiales que permiten sobrecargar operadores y definir comportamientos.

| **Método** | **Descripción** | **Ejemplo** |
| --- | --- | --- |
| `__init__(self, ...)` | Constructor | `p = Perro("Rex")` |
| `__str__(self)` | Representación informal (para `print`) | `print(p)` |
| `__repr__(self)` | Representación formal (para depuración) | `repr(p)` |
| `__len__(self)` | Para `len()` | `len(objeto)` |
| `__getitem__(self, key)` | Para acceso con corchetes | `objeto[key]` |
| `__setitem__(self, key, value)` | Asignación con corchetes | `objeto[key] = valor` |
| `__call__(self, ...)` | Hace que el objeto sea llamable como función | `objeto()` |
| `__eq__(self, other)` | Igualdad (`==`) | `obj1 == obj2` |
| `__lt__(self, other)` | Menor que (`<`) | `obj1 < obj2` |
| `__add__(self, other)` | Suma (`+`) | `obj1 + obj2` |
| `__enter__, __exit__` | Context managers (`with`) | `with objeto:` |

**Ejemplo:**

```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Punto({self.x},{self.y})"

    def __add__(self, otro):
        return Punto(self.x + otro.x, self.y + otro.y)

    def __eq__(self, otro):
        return self.x == otro.x and self.y == otro.y

p1 = Punto(1, 2)
p2 = Punto(3, 4)
print(p1 + p2)   # Punto(4, 6)
print(p1 == Punto(1, 2))  # True
```

### **8. Propiedades (`@property`)**

Permiten definir métodos que se acceden como atributos, con lógica de control.

```python
class Persona:
    def __init__(self, nombre):
        self._nombre = nombre

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, valor):
        if not valor:
            raise ValueError("El nombre no puede estar vacío")
        self._nombre = valor

    @nombre.deleter
    def nombre(self):
        print("Borrando nombre")
        del self._nombre

p = Persona("Ana")
print(p.nombre)   # Ana (accede como atributo)
p.nombre = "Luisa"  # Usa el setter
del p.nombre       # Usa el deleter
```

### **9. Encapsulación (convenciones)**

Python no tiene modificadores de acceso estrictos como Java. Se usa convención:

- `_atributo` : “protegido” (indicación de que no debería accederse desde fuera, pero es posible).
- `__atributo` : “privado” (name mangling: se renombra internamente a `_Clase__atributo` para evitar conflictos).

```python
class Ejemplo:
    def __init__(self):
        self.publico = "cualquiera"
        self._protegido = "solo por convención"
        self.__privado = "mangling"

e = Ejemplo()
print(e.publico)           # ok
print(e._protegido)        # ok (pero no debería)
print(e._Ejemplo__privado) # ok (pero no recomendado)

print(e.__privado)       # AttributeError
```

### **10. Clases abstractas**

Se usa el módulo `abc` (Abstract Base Classes) para definir interfaces o clases base abstractas.

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def hacer_sonido(self):
        pass

class Perro(Animal):
    def hacer_sonido(self):
        return "Guau"
    

a= Perro()
print(a.hacer_sonido())  # Guau
b = Animal()  # Error: no se puede instanciar clase abstracta
```

### **11. Dataclasses (Python 3.7+)**

Simplifican la creación de clases que principalmente almacenan datos.

```python
from dataclasses import dataclass

@dataclass
class Persona:
    nombre: str     # aquí hay que usar type hints
    edad: int = 0   # Valor por defecto

p = Persona("Ana", 30)
print(p.nombre)   # Ana
# Se genera automáticamente __init__, __repr__, __eq__, etc.
```

### **13. Ejemplo integrador**

```python
from abc import ABC, abstractmethod
from datetime import date

class Vehiculo(ABC):
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self._kilometraje = 0

    @abstractmethod
    def mover(self):
        pass

    def agregar_kilometros(self, km):
        if km < 0:
            raise ValueError("No se pueden agregar km negativos")
        self._kilometraje += km

    @property
    def kilometraje(self):
        return self._kilometraje

    def __str__(self):
        return f"{self.marca}{self.modelo} ({self.kilometraje} km)"

class Coche(Vehiculo):
    def __init__(self, marca, modelo, num_puertas):
        super().__init__(marca, modelo)
        self.num_puertas = num_puertas

    def mover(self):
        return "El coche está conduciendo"

    def __str__(self):
        return super().__str__() + f",{self.num_puertas} puertas"

# Uso
mi_coche = Coche("Toyota", "Corolla", 4)
mi_coche.agregar_kilometros(100)
print(mi_coche.mover())       # El coche está conduciendo
print(mi_coche)                # Toyota Corolla (100 km), 4 puertas
```
