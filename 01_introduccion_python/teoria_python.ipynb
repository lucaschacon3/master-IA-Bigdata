### 1. Introducción y Funcionamiento Interno

### Lenguaje Interpretado, de Alto Nivel y Orientado a Objetos

Python es un lenguaje de **alto nivel** porque abstrae al programador de los detalles complejos del hardware y la gestión de memoria. Es **orientado a objetos** (casi todo son objetos en Python). Y es **interpretado**, lo que significa que el código se ejecuta línea por línea por un intérprete, en lugar de compilarse directamente a código máquina como C o C++.

### El Proceso: Código Fuente → Bytecode → CPython

Aunque Python es interpretado, internamente realiza un proceso de compilación para optimizar la ejecución.

1. **Código Fuente**: El programador escribe el código en un archivo `.py`.
2. **Análisis y AST (Abstract Syntax Tree)**: El intérprete (CPython) lee el código fuente y lo convierte en una estructura jerárquica llamada **Árbol de Sintaxis Abstracta**. Este paso verifica que la sintaxis sea correcta .
3. **Compilación a Bytecode**: El AST se compila a un lenguaje de más bajo nivel llamado **bytecode**. Este bytecode es un conjunto de instrucciones independientes de la plataforma y está diseñado para ser ejecutado de manera eficiente por una máquina virtual. Normalmente se guarda en archivos `.pyc` dentro de una carpeta `__pycache__` para acelerar ejecuciones futuras .
4. **Máquina Virtual (PVM)**: El bytecode se envía a la máquina virtual de Python, que es un gran bucle que itera sobre las instrucciones del bytecode y las ejecuta una por una. CPython actúa como una capa que traduce las instrucciones del bytecode a instrucciones que el sistema operativo y el hardware puedan entender.

### Compatibilidad con C y C++

Python y C/C++ tienen una relación muy simbiótica a través de la **API Python/C**. Esta API permite dos cosas fundamentales :

- **Escribir Módulos de Extensión**: Puedes escribir funciones y tipos en C o C++ que luego se pueden importar y usar en Python como si fueran módulos nativos. Esto es para optimizar partes del código (cuellos de botella) o para permitir que Python interactúe con librerías y hardware escritos en C/C++ (ej. librerías de gráficos, controladores de dispositivos).
- **Integrar Python en una Aplicación**: Puedes incrustar el intérprete de Python dentro de una aplicación más grande escrita en C/C++. Esto permite que los usuarios extiendan la funcionalidad de la aplicación mediante scripts de Python (ej. software de modelado 3D como Blender, videojuegos).

### Tipado Dinámico y fuerte

Es tipado dinámico porque el tipo de una variable no es necesario declararlo explícitamente, el tipo de la variable será igual al tipo del objeto que apunte. Una misma variable puede cambiar de tipo a lo largo del programa. Esto contrasta con lenguajes de tipado estático como Java, donde el tipo de una variable se declara y no puede cambiar.

```python
mi_variable = 10       # type(mi_variable) --> int
mi_variable = "Hola"   # type(mi_variable) --> str
```

Es de tipado fuerte, porque prohíbe el uso de variables como tipos de datos distintos a los que fueron definidos. A diferencia del tipado débil (ej. JavaScript, PHP), donde `5 + "5"` puede resultar en `"55"` .

```python
print("hola" + 10)    # Error
```



### **Crear y Activar un Entorno Virtual**

Un entorno virtual es una caja aislada para tu proyecto. Evita que las librerías de un proyecto interfieran con las de otro.

1. **Abrir terminal**
2. **Crear el entorno**
    
    ```
    python3 -m venv .venv
    ```
    
    Esto creará una carpeta llamada `.venv` que contendrá tu entorno virtual . Puedes llamarla como quieras pero esta es la forma normalizada, el “.” es para que sea carpeta oculta
    
3. **Activar el entorno**:
    
    ```
    source venv/bin/activate
    ```
    
    Verás que el nombre del entorno (`(venv)`) aparece al principio de la línea en tu terminal, lo que indica que está activado .
    

### Python 2 (pip) vs Python 3 (pip3)

Python 2 dejó de tener soporte oficial. Siempre se debe usar Python 3.

**El Comando `python` y `python3`**

En muchos sistemas, especialmente Linux y macOS, pueden coexistir ambas versiones.

- `python`: Suele estar enlazado a `python3`, en ocasiones podría apuntar a Python 2.
- `python3`: Apunta a la versión 3 del intérprete. Es la forma más segura y recomendada de ejecutar scripts Python.

**El Comando `pip` y `pip3`**

- `pip`: En algunos sistemas, puede estar asociado a Python 2.
- `pip3`: Asociado explícitamente a Python 3. Es el que se debe usar para instalar paquetes para Python 3.

Aunque se pueda usar el comando `pip <comando>` o `pip3 <comando>` lo recomendable es usarlo así:

```bash
python3 -m pip --version  # Este referencia a pip3 porque usamos python3 de comando
```

- **Comandos comunes**:
    
    ```bash
    python3 -m pip install <nombre_paquete>    # Instalar un paquete
    python3 -m pip uninstall <nombre_paquete>  # Desinstalar
    python3 -m pip list                        # Listar paquetes instalados
    python3 -m pip list outdated               # Listar paquetes con versiones antiguas
    python3 -m pip freeze                      # Mostrar paquetes en formato requirements
    python3 -m pip freeze > requirements.txt   # Vuelca los paquetes en requirements.txt
    python3 -m pip install -r requirements.txt # Instalar los paquetes de un archivo
    ```
    
    La “-m” es de módulo y la “-r” de requirements.

### 5. Módulos y Paquetes

### ¿Qué son?

- **Módulo**: Es cualquier archivo `.py` que contiene definiciones y sentencias de Python.
- **Paquete**: Es una forma de estructurar varios módulos en un directorio. En el sistema de archivos, un paquete es una **carpeta** que contiene un archivo especial llamado `__init__.py` más módulos o subpaquetes.

### El archivo `__init__.py`

Este archivo le indica a Python que el directorio que lo contiene debe ser tratado como un paquete. Puede estar vacío, pero su presencia es obligatoria (en versiones antiguas, pero mejor ponerlo) para que Python reconozca la carpeta como un paquete. También puede contener código de inicialización del paquete o definir la variable `__all__` para controlar qué módulos se importan cuando se usa `from paquete import *`.

```bash
utilidades/
├── __init__.py
├── archivos.py
└── redes.py
```

```python
# utilidades/__init__.py

from .archivos import leer_archivo, escribir_archivo, _procesar_interno
from .redes import descargar_url, _conexion

# Definimos qué estará disponible con "from utilidades import *"
__all__ = ['leer_archivo', 'escribir_archivo', 'descargar_url']

# También podríamos incluir variables globales
VERSION = "1.0.0"
```

### Importaciones

```python
# Importar un módulo completo
import math
print(math.pi)

# Importar objetos específicos de un módulo
from math import pi, sqrt
print(pi)

# Importar con un alias para simplificar
import numpy as np
from mi_paquete import mi_modulo as mm

# Importar todo (no recomendado en código grande, puede causar conflictos)
from math import *

# Importar un módulo de un subpaquete
import mi_paquete.subpaquete.modulo
```

### Creación e Importación de Paquetes Propios

1. Crea una estructura de carpetas:
    
    ```
    mi_paquete/
    ├── __init__.py
    ├── modulo_principal.py
    └── subpaquete/
        ├── __init__.py
        └── modulo_ayuda.py
    ```
    
2. Para usar tu paquete en un script que esté al mismo nivel que la carpeta `mi_paquete`:
    
    ```python
    # script.py
    from mi_paquete import modulo_principal
    from mi_paquete.subpaquete import modulo_ayuda
    ```

### Variables: Constantes, Locales y Globales

- **Variables**: Se crean al asignarles un valor. No necesitan declaración.
    
    ```python
    nombre = "Ana"
    nombre: str = "Ana" # Se puede usar type hint
    ```
    
- **Constantes**: Por convención (no hay una regla estricta), se nombran en mayúsculas para indicar que no deberían cambiar.
    
    ```python
    PI = 3.1416
    ```
    
- **Variables Locales**: Se definen dentro de una función y solo existen allí.
- **Variables Globales**: Se definen fuera de cualquier función. Se pueden **leer** desde una función pero para **modificarlas**, se usa la palabra clave `global`.
    
    ```python
    contador = 10
    
    def incrementar():
        global contador
        contador += 1
        print(f"Dentro de la función:{contador}")
    ```
    

### La función `main`

A diferencia de Java, Python no requiere un método `main` para ejecutarse. El código se ejecuta de arriba a abajo. Sin embargo, es una práctica común definir una función `main` para controlar el flujo del programa:

```python
def main():
    print("¡Hola, mundo!")

if __name__ == "__main__":
    main()
```

`if __name__ == "__main__"` verifica si el script se está ejecutando directamente (no importado como módulo) y, en ese caso, llama a `main()`.

### Declaración y Tipado (Dinámico vs Estático)

Python es de **tipado dinámico**. Aunque desde Python 3.6+ se puede usar **type hints** (anotaciones de tipo), estas son solo informativas y no se imponen en tiempo de ejecución.

```python
# Sin type hint (tradicional)
def saludar(nombre):
    return "Hola " + nombre

# Con type hint (mejor para legibilidad y herramientas de análisis)
def saludar(nombre: str) -> str:
    return "Hola " + nombre
```

### Tipos Básicos y Operadores

- **Tipos**:
    - `int` : Números enteros.
    - `float` : Números decimales, son con “.” no con “,”.
    - `str` : Se usan con `''` o con`""` , son inmutables (Ej. cuando concatenas es un nuevo objeto).
    - `bool`: Los valores pueden ser `True` o `False` .
    - `None` : Valores nulos, para comprobar si una variable es None se usa `is` -> `if x is None:` .
    - Para castear
        
        ```python
        variable = 1.0
        # Castaea cuando tiene sentido, int("Hola") -> Error
        str(variable)
        int(variable)
        float(variable)
        
        # Para saber el tipo:
        print(type(variable))
        ```
        
- **Operadores**:
    - Aritméticos (`+`, , , `/`, `//`, `%`, `*`)
    - Comparación (`==`, `!=`, `<`, `>`, `<=`, `>=`): Ej. comparar cadenas y ordenar alfabéticamente, no como en Java.
    - Lógicos (`and`, `or`, `not`): No se usan caracteres como `&` o `|`.
    - Asignación (`=`, `+=`, etc.): El concepto `variable++`, en Python no existe.
    - Pertenencia (`in`, `not in`): Ej. Para listas, como el `.contains()` en Java.
    - Identidad (`is`, `is not`): Compara el objeto, no el valor.

### Excepciones

Son errores que ocurren durante la ejecución del programa y que interrumpen el flujo normal. Python proporciona mecanismos para “capturar” y manejar estos errores de forma controlada.

### **Bloques `try`, `except`, `else`, `finally`**

```python
try:
    # Código que puede lanzar una excepción
    numero = int(input("Introduce un número: "))
    resultado = 10 / numero
    print(f"Resultado:{resultado}")
except ValueError:
    # Se ejecuta si ocurre un error de valor (ej. letra en lugar de número)
    print("Error: Debes introducir un número válido.")
except ZeroDivisionError:
    # Se ejecuta si se intenta dividir por cero
    print("Error: No se puede dividir entre cero.")
except Exception as e:
    # Captura cualquier otra excepción (genérico, no muy recomendado)
    print(f"Error inesperado:{e}")
else:
    # Se ejecuta si NO ocurrió ninguna excepción en el bloque try
    print("Todo funcionó correctamente.")
finally:
    # Se ejecuta SIEMPRE (haya o no excepción)
    print("Bloque finally: limpieza o cierre de recursos.")
```

### **Excepciones comunes**

- `ValueError`: cuando una función recibe un argumento de tipo correcto pero valor inapropiado.
- `TypeError`: operación aplicada a objeto de tipo incorrecto.
- `IndexError`: índice fuera de rango (listas, tuplas).
- `KeyError`: clave no encontrada en diccionario.
- `FileNotFoundError`: archivo no encontrado.
- `ZeroDivisionError`: división por cero.
- `ImportError`: fallo al importar un módulo.

### **Lanzar excepciones con `raise`**

Puedes provocar intencionadamente una excepción usando `raise`.

```python
def dividir(a, b):
    if b == 0:
        raise ValueError("El divisor no puede ser cero")
    return a / b

try:
    resultado = dividir(10, 0)
except ValueError as e:
    print(e)
```

### **Crear excepciones personalizadas**

Heredando de `Exception` (o de cualquier subclase).

```python
class SaldoInsuficienteError(Exception):
    """Excepción lanzada cuando no hay suficiente saldo."""
    pass

def retirar_dinero(saldo, cantidad):
    if cantidad > saldo:
        raise SaldoInsuficienteError(f"Saldo insuficiente:{saldo} <{cantidad}")
    return saldo - cantidad
```

### Clases y Objetos

### **1. Definición básica de una clase**

Una clase es una plantilla para crear objetos. En Python, se define con la palabra clave `class`.

```python
class Perro:
    # Variable de clase (compartida por todas las instancias)
    especie = "Canis familiaris"

    def __init__(self, nombre, edad):
        # Variables de instancia
        self.nombre = nombre
        self.edad = edad

    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau!")
```

### **Características**

- **Variables de clase**: se definen fuera de cualquier método y son compartidas por todas las instancias.
- **`__init__`**: es el método constructor (inicializador). Se llama automáticamente al crear una instancia. El primer parámetro es `self`, que referencia a la instancia actual.
- **Variables de instancia**: se crean con `self.nombre`, son propias de cada objeto.
- **Métodos**: funciones definidas dentro de la clase; el primer parámetro es `self`.

### **Creación de objetos**

```python
mi_perro = Perro("Rex", 5)   # Se llama a __init__
print(mi_perro.nombre)       # Rex
mi_perro.ladrar()            # Rex dice: ¡Guau!
```

### **2. Comparación con Java**

| **Concepto** | **Python** | **Java** |
| --- | --- | --- |
| **Definición de clase** | `class Perro:` | `public class Perro { ... }` |
| **Constructor** | `def __init__(self, ...):` | `public Perro(...) { ... }` |
| **Referencia a la instancia** | `self` (explícito, primer parámetro) | `this` (implícito, se puede usar) |
| **Variables de clase** | Definidas directamente en la clase | `static` |
| **Variables de instancia** | Creadas con `self.atributo` | Declaradas en la clase, inicializadas en constructor |
| **Métodos** | `def metodo(self):` | `public void metodo() { ... }` |
| **Creación de objeto** | `obj = Clase()` | `Clase obj = new Clase();` |
| **Modificadores de acceso** | Convención: `_protected`, `__private` (name mangling) | `public`, `protected`, `private` |

### **3. El parámetro `self`**

- `self` es una referencia a la instancia actual. No es una palabra reservada (podrías llamarlo `this`), pero **por convención se usa `self`**.
- En Java, `this` es implícito; en Python hay que declararlo explícitamente como primer parámetro.

```python
class Ejemplo:
    def metodo(self, x):
        self.x = x   # Asigna a la instancia
```

### **4. Variables de clase vs de instancia**

```python
class Perro:
    especie = "Canis familiaris"   # Variable de clase

    def __init__(self, nombre):
        self.nombre = nombre        # Variable de instancia

# Acceso
print(Perro.especie)                # Desde la clase
print(mi_perro.especie)             # Desde la instancia (hereda el valor)
```

### **5. Métodos de instancia, de clase y estáticos**

Python ofrece tres tipos de métodos:

```python
class Ejemplo:
    def metodo_instancia(self):
        return "Instancia"

    @classmethod
    def metodo_clase(cls):
        return "Clase"

    @staticmethod
    def metodo_estatico():
        return "Estático"
```

| **Tipo** | **Decorador** | **Primer parámetro** | **Acceso** |
| --- | --- | --- | --- |
| **Instancia** | ninguno | `self` | Puede acceder a la instancia y a la clase |
| **Clase** | `@classmethod` | `cls` | Puede acceder solo a la clase (no a la instancia) |
| **Estático** | `@staticmethod` | ninguno | No accede ni a instancia ni a clase (función dentro de la clase) |

python

```python
e = Ejemplo()
print(e.metodo_instancia())   # Instancia
print(Ejemplo.metodo_clase()) # Clase (también desde instancia)
print(Ejemplo.metodo_estatico()) # Estático
```

### **7. Métodos especiales (mágicos)**

Los métodos que empiezan y terminan con doble guion bajo (`__` ) son métodos especiales que permiten sobrecargar operadores y definir comportamientos.

| **Método** | **Descripción** | **Ejemplo** |
| --- | --- | --- |
| `__init__(self, ...)` | Constructor | `p = Perro("Rex")` |
| `__str__(self)` | Representación informal (para `print`) | `print(p)` |
| `__repr__(self)` | Representación formal (para depuración) | `repr(p)` |
| `__len__(self)` | Para `len()` | `len(objeto)` |
| `__getitem__(self, key)` | Para acceso con corchetes | `objeto[key]` |
| `__setitem__(self, key, value)` | Asignación con corchetes | `objeto[key] = valor` |
| `__call__(self, ...)` | Hace que el objeto sea llamable como función | `objeto()` |
| `__eq__(self, other)` | Igualdad (`==`) | `obj1 == obj2` |
| `__lt__(self, other)` | Menor que (`<`) | `obj1 < obj2` |
| `__add__(self, other)` | Suma (`+`) | `obj1 + obj2` |
| `__enter__, __exit__` | Context managers (`with`) | `with objeto:` |

**Ejemplo:**

```python
class Punto:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Punto({self.x},{self.y})"

    def __add__(self, otro):
        return Punto(self.x + otro.x, self.y + otro.y)

    def __eq__(self, otro):
        return self.x == otro.x and self.y == otro.y

p1 = Punto(1, 2)
p2 = Punto(3, 4)
print(p1 + p2)   # Punto(4, 6)
print(p1 == Punto(1, 2))  # True
```

### **8. Propiedades (`@property`)**

Permiten definir métodos que se acceden como atributos, con lógica de control.

```python
class Persona:
    def __init__(self, nombre):
        self._nombre = nombre

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, valor):
        if not valor:
            raise ValueError("El nombre no puede estar vacío")
        self._nombre = valor

    @nombre.deleter
    def nombre(self):
        print("Borrando nombre")
        del self._nombre

p = Persona("Ana")
print(p.nombre)   # Ana (accede como atributo)
p.nombre = "Luisa"  # Usa el setter
del p.nombre       # Usa el deleter
```

### **9. Encapsulación (convenciones)**

Python no tiene modificadores de acceso estrictos como Java. Se usa convención:

- `_atributo` : “protegido” (indicación de que no debería accederse desde fuera, pero es posible).
- `__atributo` : “privado” (name mangling: se renombra internamente a `_Clase__atributo` para evitar conflictos).

```python
class Ejemplo:
    def __init__(self):
        self.publico = "cualquiera"
        self._protegido = "solo por convención"
        self.__privado = "mangling"

e = Ejemplo()
print(e.publico)           # ok
print(e._protegido)        # ok (pero no debería)
# print(e.__privado)       # AttributeError
print(e._Ejemplo__privado) # ok (pero no recomendado)
```

### **13. Ejemplo integrador**

python

```python
from abc import ABC, abstractmethod
from datetime import date

class Vehiculo(ABC):
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self._kilometraje = 0

    @abstractmethod
    def mover(self):
        pass

    def agregar_kilometros(self, km):
        if km < 0:
            raise ValueError("No se pueden agregar km negativos")
        self._kilometraje += km

    @property
    def kilometraje(self):
        return self._kilometraje

    def __str__(self):
        return f"{self.marca}{self.modelo} ({self.kilometraje} km)"

class Coche(Vehiculo):
    def __init__(self, marca, modelo, num_puertas):
        super().__init__(marca, modelo)
        self.num_puertas = num_puertas

    def mover(self):
        return "El coche está conduciendo"

    def __str__(self):
        return super().__str__() + f",{self.num_puertas} puertas"

# Uso
mi_coche = Coche("Toyota", "Corolla", 4)
mi_coche.agregar_kilometros(100)
print(mi_coche.mover())       # El coche está conduciendo
print(mi_coche)                # Toyota Corolla (100 km), 4 puertas
```

### Listas, Tuplas, Diccionarios y Sets

### 1. Listas (`list`)

En python no existen arrays, son listas dinámicas.

### Características

- **Mutables**: se pueden modificar (añadir, eliminar, cambiar elementos).
- **Ordenadas**: mantienen el orden de inserción.
- **Indexadas**: acceso por índice (0-based).
- **Heterogéneas**: pueden contener elementos de distintos tipos (aunque en la práctica suelen ser homogéneas).
- **Dinámicas**: crecen y encogen según necesidad.
- **Implementación interna**: array dinámico (similar a `ArrayList` de Java).

### Métodos principales

| Método | Descripción |
| --- | --- |
| `append(x)` | Añade `x` al final. |
| `extend(iterable)` | Añade todos los elementos de un iterable. |
| `insert(i, x)` | Inserta `x` en la posición `i`. |
| `remove(x)` | Elimina la primera aparición de `x` (lanza `ValueError` si no existe). |
| `pop([i])` | Elimina y devuelve el elemento en `i` (por defecto el último). |
| `clear()` | Vacía la lista. |
| `index(x[, start[, end]])` | Índice de la primera aparición de `x`. |
| `count(x)` | Número de apariciones de `x`. |
| `sort(key=None, reverse=False)` | Ordena la lista in-place. |
| `reverse()` | Invierte la lista in-place. |
| `copy()` | Devuelve una copia superficial. |

### Ejemplo

```python
frutas = ["manzana", "banana", "cereza"]
frutas.append("naranja")         # ["manzana", "banana", "cereza", "naranja"]
frutas.insert(1, "fresa")        # ["manzana", "fresa", "banana", "cereza", "naranja"]
frutas.remove("banana")          # ["manzana", "fresa", "cereza", "naranja"]
ultima = frutas.pop()            # "naranja", frutas queda ["manzana", "fresa", "cereza"]
frutas.sort()                    # ["cereza", "fresa", "manzana"]
print(frutas[0])                 # "cereza"
```

### Comparación con Java

| Python (`list`) | Java | Observaciones |
| --- | --- | --- |
| `append(x)` | `add(x)` | Añadir al final. |
| `extend(iterable)` | `addAll(coleccion)` | Añadir varios. |
| `insert(i, x)` | `add(i, x)` | Insertar en posición. |
| `remove(x)` | `remove(Object)` | Eliminar por valor. |
| `pop(i)` | `remove(i)` | Eliminar por índice (en Java devuelve el elemento eliminado). |
| `index(x)` | `indexOf(x)` | Buscar índice. |
| `sort()` | `Collections.sort(lista)` | Ordenar. |
| `reverse()` | `Collections.reverse(lista)` | Invertir. |
| `len(lista)` | `lista.size()` | Tamaño. |

**Nota**: En Python, las listas son más flexibles (heterogéneas) y tienen métodos integrados; en Java, se usan colecciones genéricas normalmente homogéneas.

### 2. Tuplas (`tuple`)

### Características

- **Inmutables**: no se pueden modificar después de creadas.
- **Ordenadas**: mantienen el orden de inserción.
- **Indexadas**: acceso por índice.
- **Heterogéneas**: pueden contener distintos tipos.
- **Implementación interna**: array fijo de referencias.
- **Usos típicos**: almacenar constantes, devolución múltiple en funciones, claves de diccionario.

### Métodos principales

Al ser inmutables, tienen pocos métodos:

| Método | Descripción |
| --- | --- |
| `count(x)` | Número de apariciones de `x`. |
| `index(x[, start[, end]])` | Índice de la primera aparición de `x`. |

Además soportan operadores como `+` (concatenación) y `*` (repetición), pero crean nuevas tuplas.

### Ejemplo

```python
punto = (3, 4)
print(punto[0])          # 3
# punto[0] = 5          # Error: TypeError
coordenadas = (10, 20, 30)
x, y, z = coordenadas    # Desempaquetado
print(x)                 # 10
```

### Comparación con Java

Java no tiene un tipo nativo equivalente a tupla. Las alternativas son:

- Usar `List.of(3, 4)` (inmutable pero no tipada).
- Crear una clase simple `record Punto(int x, int y) {}` (Java 14+).
- Usar bibliotecas como `javafx.util.Pair` o `org.apache.commons.lang3.tuple.Pair`.

La diferencia clave es que en Python las tuplas son una estructura liviana y muy utilizada; en Java, no hay un tipo estándar tan simple.

### 3. Diccionarios (`dict`)

### Características

- **Mutables**: se pueden añadir, modificar y eliminar pares clave-valor.
- **No ordenados**: históricamente, pero desde Python 3.7 **mantienen el orden de inserción** (es un detalle de implementación que pasó a ser garantizado).
- **Claves únicas**: no puede haber claves duplicadas.
- **Claves deben ser hashables** (inmutables: números, strings, tuplas, etc.).
- **Acceso por clave**, no por índice.
- **Implementación interna**: tabla hash (similar a `HashMap` de Java).
- **Usos típicos**: Usado para traducir de JSON a Python.

### Métodos principales

| Método | Descripción |
| --- | --- |
| `keys()` | Vista de las claves. |
| `values()` | Vista de los valores. |
| `items()` | Vista de pares (clave, valor). |
| `get(k[, default])` | Devuelve el valor de `k` o `default` si no existe (evita `KeyError`). |
| `update([otro])` | Actualiza con pares de otro diccionario o iterable. |
| `pop(k[, default])` | Elimina y devuelve el valor de `k`. |
| `popitem()` | Elimina y devuelve el último par insertado (LIFO). |
| `clear()` | Vacía el diccionario. |
| `setdefault(k[, default])` | Si `k` existe, devuelve su valor; si no, inserta `k` con `default` y lo devuelve. |
| `copy()` | Copia superficial. |

### Ejemplo

```python
persona = {"nombre": "Ana", "edad": 30}
print(persona["nombre"])          # "Ana"
persona["ciudad"] = "Madrid"      # Añadir
edad = persona.get("edad", 0)     # 30
persona.update({"profesion": "Ingeniera", "edad": 31})
del persona["ciudad"]             # Eliminar clave
for k, v in persona.items():
    print(k, v)
```

### Comparación con Java

| Python (`dict`) | Java | Observaciones |
| --- | --- | --- |
| `d["clave"]` | `map.get("clave")` | Acceso (en Python lanza `KeyError` si no existe; en Java devuelve `null`). |
| `d.get("clave", default)` | `map.getOrDefault("clave", default)` | Valor por defecto. |
| `d.keys()` | `map.keySet()` | Conjunto de claves. |
| `d.values()` | `map.values()` | Colección de valores. |
| `d.items()` | `map.entrySet()` | Conjunto de entradas. |
| `d["clave"] = valor` | `map.put("clave", valor)` | Insertar/modificar. |
| `del d["clave"]` | `map.remove("clave")` | Eliminar. |
| `"clave" in d` | `map.containsKey("clave")` | Comprobar existencia. |
| `len(d)` | `map.size()` | Tamaño. |

**Nota**: En Python, los diccionarios son omnipresentes y muy optimizados. En Java, `HashMap` es similar pero con genéricos y métodos específicos.

### 4. Conjuntos (`set`)

### Características

- **Mutables**: se pueden añadir y eliminar elementos.
- **No ordenados**: no garantizan orden (aunque en CPython los enteros pequeños pueden aparecer ordenados por azar de implementación, no hay que confiar).
- **Sin duplicados**: cada elemento es único.
- **Elementos deben ser hashables** (inmutables).
- **Implementación interna**: tabla hash (similar a `HashSet` de Java, que es un `HashMap` con valores dummy).
- Usos comunes: eliminar duplicados automáticamente de una lista, operaciones de unión, intersección o diferencia entre grupos de datos.

### Métodos principales

| Método | Descripción |
| --- | --- |
| `add(x)` | Añade `x`. |
| `remove(x)` | Elimina `x` (lanza `KeyError` si no existe). |
| `discard(x)` | Elimina `x` si existe (no lanza error). |
| `pop()` | Elimina y devuelve un elemento arbitrario. |
| `clear()` | Vacía el conjunto. |
| `copy()` | Copia superficial. |
| `union(otro)` | Devuelve un nuevo conjunto con elementos de ambos (también `\|`). |
| `intersection(otro)` | Intersección (también `&`). |
| `difference(otro)` | Diferencia (también `-`). |
| `symmetric_difference(otro)` | Diferencia simétrica (también `^`). |
| `issubset(otro)` | Subconjunto (`<=`). |
| `issuperset(otro)` | Superconjunto (`>=`). |
| `isdisjoint(otro)` | Si no tienen elementos en común. |

### Ejemplo

```python
colores = {"rojo", "verde", "azul"}
colores.add("amarillo")          # {"rojo", "verde", "azul", "amarillo"}
colores.remove("verde")          # {"rojo", "azul", "amarillo"}
print("rojo" in colores)         # True
primarios = {"rojo", "azul", "amarillo"}
print(colores.issubset(primarios))  # False (falta "amarillo"? realmente sí, pero colores tiene "amarillo"? Espera, revisemos: colores después de remove es {"rojo", "azul", "amarillo"} -> es subconjunto de primarios? Sí, primarios tiene esos tres, entonces es True.
```

### Comparación con Java

| Python (`set`) | Java | Observaciones |
| --- | --- | --- |
| `add(x)` | `add(x)` | Añadir. |
| `remove(x)` | `remove(x)` | Eliminar. |
| `discard(x)` | No hay directo | En Java, `remove` lanza excepción si no existe, pero puedes comprobar antes con `contains`. |
| `pop()` | No hay equivalente directo | En Java, puedes iterar y eliminar, pero no hay un “pop” arbitrario. |
| `union(otro)` | `addAll(otro)` | Unión (modifica el conjunto en Java). |
| `intersection(otro)` | `retainAll(otro)` | Intersección (modifica en Java). |
| `difference(otro)` | `removeAll(otro)` | Diferencia (modifica en Java). |
| `x in s` | `s.contains(x)` | Pertenencia. |
| `len(s)` | `s.size()` | Tamaño. |

**Nota**: En Python, los sets soportan operadores matemáticos (`\\|`, `&`, `-`, `^`) que devuelven nuevos conjuntos, mientras que en Java los métodos (`addAll`, `retainAll`, etc.) suelen modificar la colección original (aunque también hay `Set` copias inmutables).

## Estructuras adicionales que también se usan

Además de las cuatro fundamentales, Python ofrece otras estructuras muy útiles, principalmente en el módulo `collections`.

### 5. `deque` (double-ended queue)

- **Colección** optimizada para inserciones y eliminaciones por ambos extremos.
- **Métodos**: `append`, `appendleft`, `pop`, `popleft`, `extend`, `extendleft`, `rotate`.
- **Útil para**: colas, pilas, buffers circulares.

```python
from collections import deque
cola = deque(["a", "b", "c"])
cola.append("d")          # por la derecha
cola.appendleft("z")      # por la izquierda
print(cola.popleft())     # "z"
```

### 6. `Counter` (subclase de dict)

- **Cuenta** elementos de un iterable.
- **Métodos**: `elements()`, `most_common([n])`, `subtract()`.
- **Útil para**: frecuencias, histogramas.

```python
from collections import Counter
cnt = Counter(["a", "b", "a", "c", "a", "b"])
print(cnt)                # Counter({'a': 3, 'b': 2, 'c': 1})
print(cnt.most_common(2)) # [('a', 3), ('b', 2)]
```

### 7. `defaultdict` (subclase de dict)

- **Diccionario** que proporciona un valor por defecto para claves no existentes.
- **Útil para**: evitar comprobaciones de existencia.

```python
from collections import defaultdict
dd = defaultdict(int)     # valor por defecto 0
dd["a"] += 1              # no hace falta inicializar
print(dd["a"])            # 1
print(dd["b"])            # 0 (se crea automáticamente)
```

### 8. `OrderedDict` (subclase de dict)

- **Mantiene el orden de inserción** (antes de Python 3.7 era necesario; ahora `dict` ya lo hace, pero `OrderedDict` tiene métodos adicionales como `move_to_end`).
- **Útil cuando necesitas reordenar pares**.

```python
from collections import OrderedDict
od = OrderedDict()
od["a"] = 1
od["b"] = 2
od.move_to_end("a")       # mueve "a" al final
```

### 9. `namedtuple` (fábrica de tuplas con nombre)

- **Crea** una clase de tupla cuyos campos tienen nombre.
- **Útil para**: estructuras ligeras e inmutables con acceso por nombre.

```python
from collections import namedtuple
Punto = namedtuple("Punto", ["x", "y"])
p = Punto(3, 4)
print(p.x, p.y)           # 3 4
```

### 10. `array` (del módulo `array`)

- **Similar a lista** pero homogénea y más eficiente en memoria para tipos básicos.
- **Tipos**: `'i'` (int), `'f'` (float), etc.
- **Útil para**: grandes cantidades de datos numéricos.

```python
from array import array
arr = array('i', [1, 2, 3])
arr.append(4)
```

### 11. `bytes` y `bytearray`

- **Secuencias de bytes** (inmutables y mutables respectivamente).
- **Útiles para**: datos binarios, protocolos de red, E/S.

```python
b = b"hola"               # bytes inmutable
ba = bytearray(b)         # mutable
ba[0] = 72                # cambia 'h' a 'H' (ASCII)
```

### 12. `frozenset`

- **Versión inmutable** de `set`.
- Útil cuando necesitas un conjunto que pueda ser clave de diccionario o elemento de otro conjunto.

```python
fs = frozenset([1, 2, 3])
# fs.add(4)  # Error: 'frozenset' object has no attribute 'add'
```

### Diferencias Clave con Java

- **Sin punto y coma**: Las sentencias no terminan en `;`.
- **Tabulación**: Los bloques deben tabularse, no usan llaves `{}`.
- **Sin tipos primitivos**: Todo es un objeto (incluso los enteros).
- **Sin `public`, `private`**: Los niveles de acceso se manejan por convención (ej. `_atributo` para “protegido”, `__atributo` para “privado”).
- **Herencia múltiple**: Python la soporta directamente, a diferencia de Java (que usa interfaces).

| Característica | Python | Java |
| --- | --- | --- |
| **Sintaxis** | Concisa, legible y fácil de aprender. Ideal para principiantes . | Verbosa y más rígida. Requiere más líneas de código para lo mismo. |
| **Tipado** | Dinámico y fuerte. El tipo se infiere en tiempo de ejecución. | Estático y fuerte. El tipo se debe declarar en tiempo de compilación. |
| **Paradigma** | Multiparadigma (Orientado a objetos, funcional, procedural) . | Principalmente orientado a objetos (aunque desde versiones recientes incluye elementos funcionales). |
| **Rendimiento** | Generalmente más lento en tareas puras de CPU debido a la interpretación. Las librerías críticas suelen estar en C. | Generalmente más rápido y eficiente para aplicaciones de servidor de gran escala. |
