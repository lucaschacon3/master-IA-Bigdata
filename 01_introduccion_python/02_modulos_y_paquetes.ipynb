### Módulos y Paquetes

### ¿Qué son?

- **Módulo**: Es cualquier archivo `.py` que contiene definiciones y sentencias de Python.
- **Paquete**: Es una forma de estructurar varios módulos en un directorio. En el sistema de archivos, un paquete es una **carpeta** que contiene un archivo especial llamado `__init__.py` más módulos o subpaquetes.

### El archivo `__init__.py`

Este archivo le indica a Python que el directorio que lo contiene debe ser tratado como un paquete. Puede estar vacío, pero su presencia es obligatoria (en versiones antiguas, pero mejor ponerlo) para que Python reconozca la carpeta como un paquete. También puede contener código de inicialización del paquete o definir la variable `__all__` para controlar qué módulos se importan cuando se usa `from paquete import *`.

```bash
utilidades/
├── __init__.py
├── archivos.py
└── redes.py
```

````python
# utilidades/__init__.py

from .archivos import leer_archivo, escribir_archivo, _procesar_interno
from .redes import descargar_url, _conexion

# Definimos qué estará disponible con "from utilidades import *"
__all__ = ['leer_archivo', 'escribir_archivo', 'descargar_url']

# También podríamos incluir variables globales
VERSION = "1.0.0"
````

* Importaciones:

````python

# Importar un módulo completo
import math
print(math.pi)

# Importar objetos específicos de un módulo
from math import pi, sqrt
print(pi)

# Importar con un alias para simplificar
import numpy as np
from mi_paquete import mi_modulo as mm

# Importar todo (no recomendado en código grande, puede causar conflictos)
from math import *

# Importar un módulo de un subpaquete
import mi_paquete.subpaquete.modulo
````
